```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/tv-shows.csv")
df.head()
```

The dataset contains information about TV shows and movies including their genre, platform, rating, description, title etc

```python
# Column names
df.columns
```

```python
# Shape of the dataset
df.shape
```

```python
# Basic information regarding about dataset like data types, total number of columns, rows etc
df.info()
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Analysing the target variable
df['listed_in'].value_counts().head(10)
```

```python
# Genre distribution plot
plt.Figure(figsize=(10,5))
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
df['duration'].value_counts()
```

```python
# Movies Vs TV Shows plot
sns.countplot(x='type',data=df)
plt.title('Movies Vs TV Shows')
plt.show()
```

```python
# Rating distribution
sns.countplot(x='rating',data=df,order=df['rating'].value_counts().index)
plt.title('Rating Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
# Platform Distribution plot
sns.countplot(x='platform',data=df)
plt.title('Content by Platform')
plt.show()
```

```python
# Content Distribution by Country
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top Countries Producing Content')
plt.show()
```

```python
# Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

Key insights from EDA:

1. Dataset contains 9338 rows and 13 columns in which some are numerical and some are categorical columns.
2. Missing values were found in columns such as director, cast, country, rating, duration and date_added.
3. There are no duplicate records present in the dataset.
4. Documetaries, Dramas and international movies are the most common genres.
5. Movies appear more frequently than TV Shows.
6. Netflix platform produces more content compared to Disney
7. Within all other countries United States are producting more content and next India.
